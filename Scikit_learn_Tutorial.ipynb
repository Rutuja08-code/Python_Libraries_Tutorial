#Value Imputation

```python
from sklearn.impute import SimpleImputer
```

```python
print(f"Number of null values before imputing:{df.age.isnull().sum()}")
```

```python
imp=SimpleImputer(strategy='mean') # Fill missing age values using mean
df['age']=imp.fit_transform(df[['age']])
print(f"Number of null values in age column after impution:{df.age.isnull().sum()}")
```

```python
def get_parameters(df):   # Decide imputation strategy for each column with missing values
  Parameters ={}
  for col in df.columns[df.isnull().any()]:
    if df[col].dtypes=='float64' or df[col].dtypes=='int64' or df[col].dtypes=='int32':
      strategy='mean'
    else:
      strategy='most_frequent'

    missing_values=df[col][df[col].isnull()].values[0]
    Parameters[col]={'missing_values':missing_values, 'strategy':strategy}
  return Parameters

get_parameters(df)
```

```python
parameters=get_parameters(df)
```

```python
for col, param in parameters.items():  # Apply appropriate imputation to each column
  missing_values =param['missing_values']
  strategy=param['strategy']
  imp = SimpleImputer(missing_values=missing_values, strategy=strategy)
  df[col] = imp.fit_transform(df[[col]]).ravel()
```
